```python
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_csv("^IXIC.csv")
df
```

```python
setter = setup(df, target = 'Close', ignore_features=['Open', 'High', 'Low', 'Adj Close', 'Volume'])
```

```python
bester = compare_models(sort='MSE', n_select=3)
```

```python
bModel = blend_models(estimator_list = bester,  fold = 10, optimize = 'MSE', turbo = True, verbose = True)
```

```python
plot_model(bModel)
```

```python
plot_model(bModel, plot='learning')
```

```python
plot_model(bModel, plot='manifold')
```

```python
plot_model(bModel, plot='error')
```
